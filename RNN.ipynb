```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
pizza_df = plt.imread('test/pizza2.jpg')/255
pizza2_df = plt.imread('test/pizza.jpg')/255
fig, ax = plt.subplots(2,2, figsize=(6,4))
ax[0,0].imshow(pizza_df)
ax[0,1].imshow(pizza2_df);
ax[1,0].imshow(tf.image.resize(pizza_df, [256,256]))
ax[1,1].imshow(tf.image.resize(pizza2_df, [256,256]));
```

It works pretty well, so we can start feeding different images with different sizes to our Neural Network.

```python
import pandas as pd
import numpy as np
import os

folder = 'pizza'
image_data = []
for file_name in os.listdir(folder):
    if file_name.endswith(".jpg"):
        image = plt.imread(os.path.join(folder, file_name)) / 255.0
        image = tf.image.resize(image, [128,128])
        image_data.append(image)
raw_df = pd.DataFrame({'Image_Data': image_data})
raw_df.head(3)
```
